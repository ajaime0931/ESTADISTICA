```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar archivo CSV
df = pd.read_csv("producioncrudo.csv")

# Información básica
print("Información del dataset")
print(df.info())
print("Primeras filas del dataset")
print(df.head())
print("Resumen estadístico del dataset")
print(df.describe())
```

```python
# Valores faltantes
print("\nValores faltantes por columna:")
print(df.isnull().sum())

# Visualización de nulos
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Mapa de datos faltantes")
plt.show()
```

```python
df_subset = df[['ProduccionBls', 'Campo', 'Municipio', 'Departamento', 'TipoContrato', 'Vigencia']]
display(df_subset.head())
```

```python
# Valores faltantes
print("\nValores faltantes por columna:")
print(df_subset.isnull().sum())

# Visualización de nulos
sns.heatmap(df_subset.isnull(), cbar=False, cmap="viridis")
plt.title("Mapa de datos faltantes")
plt.show()
```

```python
df_subset = df_subset.dropna(subset=['TipoContrato'])

# Verify the missing values have been removed
print("\nValores faltantes por columna después de eliminar nulos:")
print(df_subset.isnull().sum())

# Display the first few rows of the cleaned subset
display(df_subset.head())
```

```python
# Visualización de nulos después de eliminar
sns.heatmap(df_subset.isnull(), cbar=False, cmap="viridis")
plt.title("Mapa de datos faltantes después de eliminar nulos")
plt.show()
```

```python
# Calculate the first quartile (Q1) and third quartile (Q3) for 'ProduccionBls'
Q1 = df_subset['ProduccionBls'].quantile(0.25)
Q3 = df_subset['ProduccionBls'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df_subset[(df_subset['ProduccionBls'] < lower_bound) | (df_subset['ProduccionBls'] > upper_bound)]

print(f"Number of outliers identified: {len(outliers)}")

# Remove outliers from the DataFrame
df_cleaned = df_subset[(df_subset['ProduccionBls'] >= lower_bound) & (df_subset['ProduccionBls'] <= upper_bound)]

print(f"Number of rows after removing outliers: {len(df_cleaned)}")

# Display the first few rows of the cleaned DataFrame
display(df_cleaned.head())
```

```python
# Create a histogram for 'ProduccionBls' after removing outliers
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['ProduccionBls'], bins=50, kde=True)
plt.title('Distribution of ProduccionBls (After Outlier Removal)')
plt.xlabel('ProduccionBls')
plt.ylabel('Frequency')
plt.show()
```

```python
from scipy import stats
import statsmodels.api as sm

# Shapiro-Wilk Test
shapiro_test = stats.shapiro(df_cleaned['ProduccionBls'])
print(f"Shapiro-Wilk Test: Statistic={shapiro_test.statistic:.4f}, p-value={shapiro_test.pvalue:.4f}")

# Kolmogorov-Smirnov Test (using Lilliefors for normality)
# Note: The standard ks_2samp is for comparing two samples.
# For testing normality against a theoretical distribution (like normal), lilliefors is more appropriate.
# However, scipy's kstest can be used with the 'norm' distribution for a single sample test.
ks_test = stats.kstest(df_cleaned['ProduccionBls'], 'norm', args=(df_cleaned['ProduccionBls'].mean(), df_cleaned['ProduccionBls'].std()))
print(f"Kolmogorov-Smirnov Test: Statistic={ks_test.statistic:.4f}, p-value={ks_test.pvalue:.4f}")

# Anderson-Darling Test
ad_test = stats.anderson(df_cleaned['ProduccionBls'], dist='norm')
print(f"Anderson-Darling Test:")
print(f" Statistic: {ad_test.statistic:.4f}")
for i in range(len(ad_test.critical_values)):
    print(f" Critical Value ({ad_test.significance_level[i]}%): {ad_test.critical_values[i]:.4f}")

# Jarque-Bera Test
jb_test = stats.jarque_bera(df_cleaned['ProduccionBls'])
print(f"Jarque-Bera Test: Statistic={jb_test.statistic:.4f}, p-value={jb_test.pvalue:.4f}")

# QQ Plot
plt.figure(figsize=(8, 6))
sm.qqplot(df_cleaned['ProduccionBls'], line='s')
plt.title('QQ Plot of ProduccionBls (After Outlier Removal)')
plt.show()
```
